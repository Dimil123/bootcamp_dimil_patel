```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from scipy.stats import skew, kurtosis
sns.set(context='talk', style='darkgrid')
np.random.seed(8)
pd.set_option('display.max_columns', 100)

# === Synthetic data generator (adapt or replace with your own data) ===
n = 200
df = pd.DataFrame({
    'date': pd.date_range('2021-02-01', periods=n, freq='D'),
    'region': np.random.choice(['North','South','East','West'], size=n),
    'age': np.random.normal(40, 8, size=n).clip(22, 70).astype(int),
    'income': np.random.lognormal(mean=10.6, sigma=0.3, size=n).round(2),
    'transactions': np.random.poisson(lam=3, size=n),
})
base = df['income'] * 0.0015 + df['transactions']*18 + np.random.normal(0, 40, size=n)
df['spend'] = np.maximum(0, base).round(2)

# inject a bit of missingness and outliers
df.loc[np.random.choice(df.index, 12, replace=False), 'income'] = np.nan
df.loc[np.random.choice(df.index, 7, replace=False), 'spend'] = np.nan
df.loc[np.random.choice(df.index, 5, replace=False), 'transactions'] = df['transactions'].max()+12
df.head()
```

## 1) First look

```python
df.info(), df.isna().sum()
```

## 2) Numeric profile

```python
desc = df[['age','income','transactions','spend']].describe().T
desc['skew'] = [skew(df[c].dropna()) for c in desc.index]
desc['kurtosis'] = [kurtosis(df[c].dropna()) for c in desc.index]
desc
```

## 3) Distributions

```python
# histogram + KDE for a key variable
sns.histplot(df['income'], kde=True)
plt.title('Income Distribution')
plt.show()

# boxplot for outliers
sns.boxplot(x=df['transactions'])
plt.title('Transactions (Outliers)')
plt.show()

# Spend Distibution
sns.histplot(df['spend'], kde=True)
plt.title('Spend Distribution')
plt.show()
```

## 4) Relationships

```python
sns.scatterplot(data=df, x='income', y='spend', hue='region')
plt.title('Income vs Spend')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()

sns.scatterplot(data=df, x='age', y='spend')
plt.title('Age vs Spend')
plt.show()

sns.lineplot(data=df, x='income',y="transactions", hue='region')
plt.title('Income wise transactions')
plt.show()
```

## 5) Correlation matrix

```python
corr = df[['age','income','transactions','spend']].corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt='.2f', cmap='PiYG', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
corr
```

## Top 3 Insights

#Income Distribution is Right-Skewed
The income variable follows a log-normal distribution, showing strong right skewness. This suggests that while most individuals have moderate incomes, a few have exceptionally high incomes, which could influence mean-based analyses.

#Spend is Correlated with Transactions and Income
As expected, spend increases with both the number of transactions and income. This aligns with the formula used to generate the synthetic spend variable and reflects realistic consumer behavior patterns.

#Presence of Outliers and Missing Data
Some unusually high values in transactions and missing values in income and spend indicate data quality issues. These outliers could distort models, and the missing values will need to be handled before further analysis.

## Assumptions & Risks

Assumption of Normality (for Age): We're assuming that age is approximately normally distributed due to the way it's generated. If applied to real data, this assumption should be validated.

Synthetic Data Reflects Reality: The data was artificially generated and may not fully represent real-world variability, especially in features like region or spending behavior.

Outliers are Noise: We are assuming that extreme values in transactions are anomalies or errors. However, they might represent real high-activity users and could contain useful signals.

Risks:

Removing true extreme behaviors could reduce the model's predictive power.

Imputing missing values incorrectly could bias the dataset.

## Next Steps Before Modeling

#Missing Value Handling:

Extend income and spend using median or regression-based replacement.

#Outlier Treatment:

Trim or remove outliers in transactions, depending on whether they're noise or signal.

#Feature Engineering:

Create new features like spend_per_transaction, or categorize income into bins.

Consider encoding region using one-hot or ordinal encoding.
